```python
def test_time(num):
    df = pd.DataFrame(columns = ['id', 'time', 'viewer_num'])
    now = datetime.now()
    
    # 지금 라이브하고 있는 방송 목록
    request = youtube.search().list(
        part="snippet",
        eventType="live",
        maxResults=num,
        # videoCategoryId=20,
        regionCode='KR',
        type="video"
    )
    response = request.execute()

    channel_list = []
    for item in response['items']:
        channel_list.append(item['snippet']['channelId'])

    # 방송 목록 서치 및 각 영상의 정보 추출
    l_ts = time.time()
    for channel in channel_list:
        request = youtube.search().list(
            part="snippet",
            eventType="live",
            maxResults=25,
            regionCode='KR',
            type="video",
            
        )
        ch_response = request.execute()
        video_id = ch_response['items'][0]['id']['videoId']

        request = youtube.videos().list(
            part="snippet,contentDetails,statistics,liveStreamingDetails",
            id = video_id
        )
        
        response = request.execute()
        item = response['items'][0]
        # concurrentViewers가 없다고 에러가 나서 추가
        # 그 뒤로는 에러가 난 적이 없어서 어떤 상황인지 아직 찾지 못했음.
        if item['liveStreamingDetails']['concurrentViewers']:
            df.loc[len(df)] = [video_id, now.strftime('%Y-%m-%d %H:%M'), item['liveStreamingDetails']['concurrentViewers']]
        else:
            print(item)

        
    l_ts = time.time() - l_ts
    print(l_ts)

    return df
```
